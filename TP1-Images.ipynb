```python
def binarize_array(numpy_array, threshold=200):
    """Binarize a numpy array."""
    for i in range(len(numpy_array)):
        for j in range(len(numpy_array[0])):
            if numpy_array[i][j] > threshold:
                numpy_array[i][j] = 255
            else:
                numpy_array[i][j] = 0
    return numpy_array


image_file = Image.open('snake.png')
image = image_file.convert('L')  # convert image to monochrome
image = numpy.array(image)
image = binarize_array(image, threshold=50)

plt.imshow(image)
```

# Projection

```python
img = mpimg.imread('./salut.jpeg')
plt.imshow(img)

```

```python
#Converting to black and white
import numpy as np
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
(thresh, blackAndWhiteImage) = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
array = np.array(blackAndWhiteImage)
print(array.shape)
print(np.unique(array))
horz_proj = [np.count_nonzero(row) for row in array]
img_trans = array.transpose()
ver_proj = [np.count_nonzero(row) for row in img_trans]

plt.plot(ver_proj)
plt.ylabel('Vertical Projection')
plt.show()
```
