```python
import numpy as np
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from matplotlib.image import imread
from sklearn.metrics import jaccard_score
from scipy import ndimage
```

```python
digits=load_digits()
```

```python
plt.gray()
plt.matshow(digits.images[0])
plt.show()
```

<h2>Distances<h2>

```python
n_images=digits.images.shape[0]
distances =np.empty((n_images,n_images))
for i in range(n_images):
    for j in range(n_images):
        distances[i][j]=np.linalg.norm(digits.images[i]-digits.images[j])
```

<h2>Similarity<h2>
    <h3>Jaccard<h3>

```python
import numpy as np
from sklearn.metrics import jaccard_score
y_true = np.array([[0, 1, 1],
                   [1, 1, 0]])
y_pred = np.array([[1, 1, 1],
                   [1, 0, 0]])
jaccard_score(y_true[0],y_pred[0])
jaccard_score(y_true, y_pred, average='samples')
jaccard_score(y_true, y_pred, average='macro')
jaccard_score(y_true, y_pred, average=None)
```

```python
catImage = imread('cat.jpg')
imgplot = plt.imshow(catImage)
```

```python
from sklearn import cluster
from sklearn.cluster import KMeans
x, y, z = catImage.shape
image_2d = catImage.reshape(x*y, z)
n=4
kmeans_cluster = cluster.KMeans(n_clusters=n)
kmeans_cluster.fit(image_2d)
cluster_centers = kmeans_cluster.cluster_centers_
cluster_labels = kmeans_cluster.labels_
plt.imshow((cluster_centers[cluster_labels].reshape(x, y, z) * 255).astype(np.uint8))
```
