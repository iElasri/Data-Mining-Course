```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import ndimage
from sklearn import cluster
dogImage = imread('dog.jpg')
plt.imshow(dogImage)
```

```python
dogImage.shape
```

```python
x, y, z = dogImage.shape
image_2d = dogImage.reshape(x*y, z)
image_2d.shape
```

```python
n=7
kmeans_cluster = cluster.KMeans(n_clusters=n)
kmeans_cluster.fit(image_2d)
cluster_centers = kmeans_cluster.cluster_centers_
cluster_labels = kmeans_cluster.labels_
```

```python
plt.imshow((cluster_centers[cluster_labels].reshape(x, y, z) * 255).astype(np.uint8))
```

```python
cluster_centers[cluster_labels].reshape(x, y, z).shape
```

```python
clusters=np.ones((n,n,3))
clusters
```

```python
for i in range(n):
    for j in range(n):
        clusters[i][i]=list(cluster_centers[i])
    print(clusters[i])
    print(cluster_centers[i])
    plt.imshow((clusters[i][cluster_labels].reshape(x, y, z) * 255).astype(np.uint8))
    plt.show()
```
